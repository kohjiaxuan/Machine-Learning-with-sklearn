```python
X = wine.drop('quality', axis = 1)
y = wine['quality']
```

```python
len(wine)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
X_train.tail(5)
```

```python
y_test.tail(5)
```

```python
y_test[3671]
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

# Random Forest Classifier

```python
iterate = [1,5,10,20,30,50,75,100,150,200,300,500]

for i in iterate:
    rfc = RandomForestClassifier(n_estimators=i)
    rfc.fit(X_train,y_train)
    pred_rfc = rfc.predict(X_test)

    print('\nRandom Forest for '+str(i)+' trees:')
    print(classification_report(y_test, pred_rfc))
    print(confusion_matrix(y_test, pred_rfc))

    #Visualise confusion matrix
    RFC_CM = pd.DataFrame(confusion_matrix(y_test, pred_rfc), ['Actual 0','Actual 1'], ['Predict 0','Predict 1'] )
    plt.figure(figsize = (6,4))
    sns.set(font_scale=1.4)
    plt.title('Confusion matrix for '+str(i)+' trees:')
    sns.heatmap(RFC_CM, annot=True, annot_kws={"size": 18}, linewidths=1, fmt='g', cmap='Blues')
```

```python
RFC_CM
```

# SVM Classifier

```python
C = [1,1.2,1.5,2]
gamma = [0.3, 0.5, 1, 1.5, 2]

for i in gamma:
    for j in C:
        clf=svm.SVC(C=j,gamma=i,kernel='rbf')
        clf.fit(X_train,y_train)
        pred_clf = clf.predict(X_test)

        print('SVM with gamma: '+str(i)+' and C: '+str(j))
        print(classification_report(y_test, pred_clf))
        print(confusion_matrix(y_test, pred_clf))

        #Visualise confusion matrix
        SVM_CM = pd.DataFrame(confusion_matrix(y_test, pred_clf), ['Actual 0','Actual 1'], ['Predict 0','Predict 1'])
        plt.figure(figsize = (6,4))
        #Important to put plt.figure in for loop as it seperates the graph into new figures
        sns.set(font_scale=1.4)
        plt.title('SVM with gamma: '+str(i)+' and C: '+str(j))
        sns.heatmap(SVM_CM, annot=True, annot_kws={"size": 18}, linewidths=1, fmt='g', cmap='Blues')
```

```python
#Default setting of SVM without any gamma or c
clf=svm.SVC()
clf.fit(X_train,y_train)
pred_clf = clf.predict(X_test)

print(classification_report(y_test, pred_clf))
print(confusion_matrix(y_test, pred_clf))

#Visualise confusion matrix
SVM_CM = pd.DataFrame(confusion_matrix(y_test, pred_clf), ['Actual 0','Actual 1'], ['Predict 0','Predict 1'] )
plt.figure(figsize = (6,4))
sns.set(font_scale=1.4)
sns.heatmap(SVM_CM, annot=True, annot_kws={"size": 18}, linewidths=1, fmt='g', cmap='Blues')

```

## Neural Network

```python
NN=[(10,10),(15,15),(20,20),(25,25),(25,15),(10,10,10),(15,15,15),(20,20,20),(25,25,25),(25,25,15),(10,10,10,10),(15,15,15,15),(20,20,20,20),(25,25,25,25),(30,30,30,30),(35,35,35,35),(35,35,35,15),(40,40,40,40),(40,40,40,15)]
for setting in NN:
    mlpc = MLPClassifier(hidden_layer_sizes=setting, max_iter=1000)
    mlpc.fit(X_train, y_train)
    pred_mlpc = mlpc.predict(X_test)
    report = classification_report(y_test,pred_mlpc, digits=3).split('     ')
    print('Classification Report for',str(setting)+'. Recall for Positive value is',report[12]+'. Average f1 score is',report[17])
    print(classification_report(y_test,pred_mlpc, digits=3))
    print(confusion_matrix(y_test,pred_mlpc))

    #Visualise confusion matrix
    MLPC_CM = pd.DataFrame(confusion_matrix(y_test, pred_mlpc), ['Actual 0', 'Actual 1'], ['Predict 0', 'Predict 1'])
    plt.figure(figsize = (6,4))
    #Important to put plt.figure in for loop as it seperates the graph into new figures
    sns.set(font_scale=1.4)#for label size
    sns.heatmap(MLPC_CM, annot=True,annot_kws={"size": 16}, cmap='Reds', linewidths=1, fmt='g')# font size
    plt.title('Confusion Matrix for '+str(setting)+'. Recall for Positive value is '+str(report[12])+'. Average f1 score is '+ str(report[17]))
```

```python
#Getting the values out from str of classification report
#Split with '     ' (tab)
#Get the 12nd and 17th cell number which is recall for 1, and avg f1-score

print(classification_report(y_test,pred_mlpc, digits=3))
for index, word in enumerate(classification_report(y_test,pred_mlpc, digits=3).split('     ')):
    print(index, word)

#Recall of 1
print(classification_report(y_test,pred_mlpc, digits=3).split('     ')[12])

#Average F1 score
print(classification_report(y_test,pred_mlpc, digits=3).split('     ')[17])
```
